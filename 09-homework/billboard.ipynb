```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.billboard.com/charts/hot-100')
doc = BeautifulSoup(response.text)
```

```python
rows = []
songs = doc.find_all(class_ = 'chart-list-item')

for song in songs:
    row = {}
    
    row['name'] = song.find(class_ = 'chart-list-item__title').text.strip()
    row['artist'] = song.find(class_ = 'chart-list-item__artist').text.strip()
    row['rank'] = song.find(class_ = 'chart-list-item__rank').text.strip()
    
    rows.append(row)
df = pd.DataFrame(rows)
df = df[['name', 'artist', 'rank']]
df
```

```python
df.to_csv('Billboard Top 100.csv',index=False)
```
